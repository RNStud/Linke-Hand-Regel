```python
#UNBEDINGT VORHER AUSFÜHREN!: pip install pyamaze
#Version mit Pfad ohne Animation
from pyamaze import maze,agent,COLOR       #vom Pymaze Modul die nötigen Funktionen importieren
m=maze(10,10)                              #m als Maze definieren, hier ist das Labyrinth 10 * 10 gross
m.CreateMaze(theme=COLOR.light)            #macht das Labyrinth mit einem hellen Hintergrund (default wäre dunkler Hintergrund)
a=agent(m,footprints=True)                 #den agent a kreieren im Labyrinth m mit einem Pfad als bewegungstracker
print(m.maze_map)                          #gibt die Maze-Map des generierten Labyrinths unten wieder
i=-1                                       #i wechselt bei r (Richtung) und h (Hand) in der Schleife zwischen den 4 Richtungen NESW
n=True                                     #der boolean n prüft ob eine der 2 spezialfälle vorgetreten ist und deaktiviert das Wechseln der Richtungen
while a.position != (1,1):                 #Schleife, welche läuft bis der Agent im Ziel (1, 1) ist
    r = 'ESWN'
    h = 'NESW'                             # die Richtung ist hier um 1 nach links verschoben
    if n==True:                            #im Normalfall ist n=True und bei jedem Durchgang wechseln somit die Richtungen
        i=i+1
    if i==4:                               #da es bei r und h nur 4 Richtungen gibt stellt diese Bedingung i auf 0 falls i zu gross wird
        i=0
    n=True
    if m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False: #prüft ob es vor dem Agenten Frei ist und links von dem Agent eine wand ist, a.position gibt die Position des agenten a in einem Tupel 
            if i==0: #Fall: Richtung des Agenten ist Osten
                while m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False: #Während diese Kondition gilt, bewegt der Agent sich nach vorne
                    a.position=(a.position[0],a.position[1]+1)
                if m.maze_map[a.position][h[i]]==True:              #Spezialfall 1: links vom Agenten hat es keine Wand mehr, der Agent geht beim Fall Osten nach Norden
                    a.position=(a.position[0]-1,a.position[1])
                    i=3
                    n=False
                    if m.maze_map[a.position][h[i]]==True:          #Spezialfall 2: Agent hat nach ausführung von Spezialfall 1 immernoch keine Wand, der Agent geht beim Fall Osten nun nach Westen
                        a.position=(a.position[0],a.position[1]-1)
                        i=2
                        n=False                                     # bei beiden Fällen wird n=False und i wird auf die Richtung des Agenten gesetzt
            elif i==2: #Fall: Richtung des Agenten ist Westen
                while m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False:
                    a.position=(a.position[0],a.position[1]-1)
                if m.maze_map[a.position][h[i]]==True:
                    a.position=(a.position[0]+1,a.position[1])
                    i=1
                    n=False
                    if m.maze_map[a.position][h[i]]==True:
                        a.position=(a.position[0],a.position[1]+1)
                        i=0
                        n=False
            elif i==3: #Fall: Richtung des Agenten ist Norden
                while m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False:
                    a.position=(a.position[0]-1,a.position[1])
                if m.maze_map[a.position][h[i]]==True:
                    a.position=(a.position[0],a.position[1]-1)
                    i=2
                    n=False
                    if m.maze_map[a.position][h[i]]==True:
                        a.position=(a.position[0]+1,a.position[1])
                        i=1
                        n=False
            elif i==1: #Fall: Richtung des Agenten ist Süden
                while m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False:
                    a.position=(a.position[0]+1,a.position[1])
                if m.maze_map[a.position][h[i]]==True:
                    a.position=(a.position[0],a.position[1]+1)
                    i=0
                    n=False
                    if m.maze_map[a.position][h[i]]==True:
                        a.position=(a.position[0]-1,a.position[1])
                        i=3
                        n=False
# bei den Restlichen Richtungen passiert das gleiche wie beim Fall Osten, nur angepasst auf die Richtung
m.run() #Öffnet die Darstellung
```

```python
#UNBEDINGT VORHER AUSFÜHREN!: pip install pyamaze
#Version ohne Pfad mit Animation
from pyamaze import maze,agent,COLOR
m=maze(10,10)
m.CreateMaze(theme=COLOR.light)
a=agent(m)
start = a.position
weg=''
print(m.maze_map)
i=-1
n=True
while a.position != (1,1):
    r = 'ESWN'
    h = 'NESW'
    if n==True:
        i=i+1
    if i==4:
        i=0
    n=True
    if m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False:
            if i==0:
                while m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False:
                    a.position=(a.position[0],a.position[1]+1)
                    weg=weg+'E'
                if m.maze_map[a.position][h[i]]==True:
                    a.position=(a.position[0]-1,a.position[1])
                    i=3
                    n=False
                    weg=weg+'N'
                    if m.maze_map[a.position][h[i]]==True:
                        a.position=(a.position[0],a.position[1]-1)
                        i=2
                        n=False
                        weg=weg+'W'
            elif i==2:
                while m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False:
                    a.position=(a.position[0],a.position[1]-1)
                    weg=weg+'W'
                if m.maze_map[a.position][h[i]]==True:
                    a.position=(a.position[0]+1,a.position[1])
                    i=1
                    n=False
                    weg=weg+'S'
                    if m.maze_map[a.position][h[i]]==True:
                        a.position=(a.position[0],a.position[1]+1)
                        i=0
                        n=False
                        weg=weg+'E'
            elif i==3:
                while m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False:
                    a.position=(a.position[0]-1,a.position[1])
                    weg=weg+'N'
                if m.maze_map[a.position][h[i]]==True:
                    a.position=(a.position[0],a.position[1]-1)
                    i=2
                    n=False
                    weg=weg+'W'
                    if m.maze_map[a.position][h[i]]==True:
                        a.position=(a.position[0]+1,a.position[1])
                        i=1
                        n=False
                        weg=weg+'S'
            elif i==1:
                while m.maze_map[a.position][r[i]]==True and m.maze_map[a.position][h[i]]==False:
                    a.position=(a.position[0]+1,a.position[1])
                    weg=weg+'S'
                if m.maze_map[a.position][h[i]]==True:
                    a.position=(a.position[0],a.position[1]+1)
                    i=0
                    n=False
                    weg=weg+'E'
                    if m.maze_map[a.position][h[i]]==True:
                        a.position=(a.position[0]-1,a.position[1])
                        i=3
                        n=False
                        weg=weg+'N'

a.position = start
m.tracePath({a:weg})
m.run()
```
